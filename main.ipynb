```python
import pandas as pd
import tkinter as tk
from tkinter import messagebox as mb
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
from matplotlib.figure import Figure
import matplotlib.ticker as ticker
import hashlib
```

```python
my_dict = {} #stores name :  balance
my_dict2 = {} #stores name : password(hashed)
```

```python
top = tk.Tk()


top.geometry("600x400")

bgcolor = "cyan"
butcolor = "#00bbff"
headcolor = "blue"

top.rowconfigure(0,weight=1)
top.columnconfigure(0,weight=1)

#----------------------------------------

def on_enter(e):
    e.widget['background'] = '#0077ff'

def on_leave(e):
    e.widget['background'] = '#00bbff'

def f0t1():
    frame0.grid_forget()
    frame1.grid(row=0,column=0,sticky='nsew')
def f1t0():
    frame1.grid_forget()
    frame0.grid(row=0,column=0,sticky='nsew')
def f2t0():
    frame2.grid_forget()
    frame0.grid(row=0,column=0,sticky='nsew')
def f0t2():
    frame0.grid_forget()
    frame2.grid(row=0,column=0,sticky='nsew')
def f1t2():
    frame1.grid_forget()
    frame2.grid(row=0,column=0,sticky='nsew')
def f3t0():
    frame3.grid_forget()
    frame0.grid(row=0,column=0,sticky='nsew')
def f2t3():
    frame2.grid_forget()
    frame3.grid(row=0,column=0,sticky='nsew')
def tf3():
    vis.lf.grid_forget()
    frame3.grid(row=0,column=0,sticky='nsew')
def sub():
    my_dict[name.get()]=int(balance.get())
    my_dict2[name.get()]=hashlib.md5(password.get().encode()).hexdigest()
    
    w1.delete(0,tk.END)
    w2.delete(0,tk.END)
    w3.delete(0,tk.END)
def sub2():
    if(my_dict2[Name.get()]==hashlib.md5(Password.get().encode()).hexdigest()):
        
        f2t3()
    else :
        e1.delete(0,tk.END)
        e2.delete(0,tk.END)
def enter():
    h = pd.read_csv(r'C:\Users\Aziz_Shameem\Documents\PyCK Project\table.csv')
    nm1 = Name.get()
    password1 = my_dict2[nm1]
    day1 = day.get()
    month1 = month.get()
    year1 = year.get()
    title1 = title.get()
    exp1 = exp.get()
    h = h.append(pd.DataFrame([[nm1,password1,day1,month1,year1,title1,exp1,my_dict[nm1]-int(exp1)]],columns=['Name','Password','Day','Month','Year','Title','Expense','Balance']),ignore_index=True)
    h.to_csv(r'C:\Users\Aziz_Shameem\Documents\PyCK Project\table.csv',index=False)
    my_dict[nm1] -= int(exp1)
    

    q1.delete(0,tk.END)
    q2.delete(0,tk.END)
    q3.delete(0,tk.END)
    q4.delete(0,tk.END)
    q5.delete(0,tk.END)

def vis():
    h = pd.read_csv(r'C:\Users\Aziz_Shameem\Documents\PyCK Project\table.csv')
    h = h[h['Name']==Name.get()]
    
    
    vis.lf = tk.Frame(top,bg=bgcolor)
    vis.lf.grid(row=0, column=0, sticky='nwes')
    tk.Button(vis.lf, text="Back", width=16,bg=butcolor, command=tf3).place(x=230, y=350)
    fig = Figure(figsize=(6,3), dpi=100, facecolor=bgcolor)
    ax = fig.add_subplot(111)
    ax.bar(h['Year'],h['Expense'])
    #ax.set_facecolor(bgcolor)
    
    for axis in [ax.xaxis, ax.yaxis]:
        axis.set_major_locator(ticker.MaxNLocator(integer=True))
    
    canvas = FigureCanvasTkAgg(fig,master=vis.lf)
    
    #lf.grid(row=0,column=0,sticky='nsew')
    canvas.get_tk_widget().pack()
    canvas.draw()
    
    
#----------------------------------------

#    ....FRAME 0....

frame0 = tk.Frame(top, bg=bgcolor)
frame0.grid(row=0,column=0,sticky='nsew')

title_label = tk.Label(frame0,text="Expnese Tracker",fg=headcolor, bg=bgcolor,font=("Arial",20)).place(x=180,y=20)
login = tk.Button(frame0, text='Add Expense',width=16, height=2, command = f0t2, bg=butcolor)
login.place(x=225,y=100)
signup = tk.Button(frame0, text='New Account', width=16, height=2, command=f0t1, bg=butcolor)
signup.place(x=225,y=180)
exitbutton = tk.Button(frame0, text='Quit', width=16, height=2,bg=butcolor, command=lambda :top.destroy)
exitbutton.place(x=225,y=260)

login.bind('<Enter>', on_enter)
login.bind('<Leave>', on_leave)
signup.bind('<Enter>', on_enter)
signup.bind('<Leave>', on_leave)
exitbutton.bind('<Enter>', on_enter)
exitbutton.bind('<Leave>', on_leave)
#    ....END....

#    ....FRAME 1....
frame1 = tk.Frame(top, bg=bgcolor)
#frame1.grid(row=0,column=0,sticky='nsew')


name = tk.StringVar()
balance = tk.StringVar()
password = tk.StringVar()

tk.Label(frame1,text="Expnese Tracker",fg=headcolor, bg=bgcolor,font=("Arial",20)).place(x=180,y=20)

tk.Label(frame1,text="Name :", bg=bgcolor, font=("Arial",13)).place(x=140,y=100)
tk.Label(frame1,text="Initial Balance :", bg=bgcolor, font=("Arial",13)).place(x=140,y=140)
tk.Label(frame1,text="Password :", bg=bgcolor,font=("Arial",13)).place(x=140,y=180)

w1 = tk.Entry(frame1,textvariable = name,width=25)
w1.place(x=260,y=100)
w2 = tk.Entry(frame1, textvariable = balance,width=25)
w2.place(x=260,y=140)
w3 = tk.Entry(frame1, textvariable = password, width=25, show="*")
w3.place(x=260, y=180)



expbutton = tk.Button(frame1, text='Expense',width=16,height=2, bg=butcolor, command=f1t2)
expbutton.place(x=140,y=240)
backbutton = tk.Button(frame1, text='Back',width=16,height=2, bg=butcolor, command = f1t0)
backbutton.place(x=300,y=240)
subBut = tk.Button(frame1, text='Submit',width=16,height=2, bg=butcolor,command=sub)
subBut.place(x=220,y=300)

expbutton.bind('<Enter>', on_enter)
expbutton.bind('<Leave>', on_leave)
backbutton.bind('<Enter>', on_enter)
backbutton.bind('<Leave>', on_leave)
subBut.bind('<Enter>', on_enter)
subBut.bind('<Leave>', on_leave)

#    ....END....

#    ....FRAME 2....
frame2 = tk.Frame(top, bg=bgcolor)
#frame2.grid(row=0,column=0,sticky='nsew')

Name = tk.StringVar()
Password = tk.StringVar()

tk.Label(frame2,text="Expnese Tracker",fg=headcolor, bg=bgcolor,font=("Arial",20)).place(x=180,y=20)

tk.Label(frame2,text="Name :", bg=bgcolor, font=("Arial",13)).place(x=140,y=100)
tk.Label(frame2,text="Password :", bg=bgcolor, font=("Arial",13)).place(x=140,y=140)

e1 = tk.Entry(frame2,textvariable = Name,width=25)
e1.place(x=260,y=100)
e2 = tk.Entry(frame2, textvariable = Password,width=25,show="*")
e2.place(x=260,y=140)


backbutton = tk.Button(frame2, text='Back',width=16,height=2, bg=butcolor, command = f2t0)
backbutton.place(x=300,y=200)
subBut = tk.Button(frame2, text='Submit',width=16,height=2, bg=butcolor,command=sub2)
subBut.place(x=140,y=200)

backbutton.bind('<Enter>', on_enter)
backbutton.bind('<Leave>', on_leave)
subBut.bind('<Enter>', on_enter)
subBut.bind('<Leave>', on_leave)


#    ....END....

#    ....FRAME 3....

frame3 = tk.Frame(top, bg=bgcolor)

nm = Name.get()
day = tk.StringVar()
month = tk.StringVar()
year = tk.StringVar()
title = tk.StringVar()
exp = tk.StringVar()

tk.Label(frame3, text='Day ',bg=bgcolor, font=("Arial",13)).place(x=140,y=60)
tk.Label(frame3, text='Month ',bg=bgcolor, font=("Arial",13)).place(x=140,y=100)
tk.Label(frame3, text='Year ',bg=bgcolor, font=("Arial",13)).place(x=140,y=140)
tk.Label(frame3, text='Title ',bg=bgcolor, font=("Arial",13)).place(x=140,y=180)
tk.Label(frame3, text='Expense ',bg=bgcolor, font=("Arial",13)).place(x=140,y=220)


q1 = tk.Entry(frame3, textvariable=day,width=25)
q1.place(x=300,y=60)
q2 = tk.Entry(frame3, textvariable=month,width=25)
q2.place(x=300,y=100)
q3 = tk.Entry(frame3, textvariable=year,width=25)
q3.place(x=300,y=140)
q4 = tk.Entry(frame3, textvariable=title,width=25)
q4.place(x=300,y=180)
q5 = tk.Entry(frame3, textvariable=exp,width=25)
q5.place(x=300,y=220)

submitBut = tk.Button(frame3, text='Submit', width=16,height=2, bg=butcolor,command = enter)
submitBut.place(x=100,y=300)
VisBut = tk.Button(frame3, text='Visualise', width=16,height=2, bg=butcolor, command = vis)
VisBut.place(x=240,y=300)
BackBut = tk.Button(frame3, text='Back', width=16,height=2, bg=butcolor,command = f3t0)
BackBut.place(x=380,y=300)

submitBut.bind('<Enter>', on_enter)
submitBut.bind('<Leave>', on_leave)
VisBut.bind('<Enter>', on_enter)
VisBut.bind('<Leave>', on_leave)
BackBut.bind('<Enter>', on_enter)
BackBut.bind('<Leave>', on_leave)


top.mainloop()  #runs the UI
```
